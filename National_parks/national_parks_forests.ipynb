```python
import pandas as pd
import geopandas as gpd
from geopandas.tools import geocode
import lxml
```

```python
data= pd.read_html('https://en.wikipedia.org/wiki/List_of_national_parks_of_India')
```

```python
data=data[1]
data
```

```python
type(data)
```

```python
data_cleaned=data.drop(columns=['Notable Features', 'Rivers and lakes inside the national park'])
data_cleaned
```

```python
#data_cleaned.to_csv('national_parks_india_before.csv',index=False)
```

```python
data_cleaned = pd.read_csv("national_parks_india_before.csv")
```

```python
for index, row in data_cleaned.iterrows():
    print (row['Name'])
    info= geocode(str(row['Name']), provider='arcgis')#(timeout=3))
    #data_cleaned.loc[int(index), 'Address']= info['address'].loc[0]
    data_cleaned.loc[int(index), 'lon']=info['geometry'].loc[0].x
    data_cleaned.loc[int(index), 'lat']=info['geometry'].loc[0].y
```

```python
data_cleaned
```

```python
data_cleaned.to_csv('national_parks_india.csv')
```
